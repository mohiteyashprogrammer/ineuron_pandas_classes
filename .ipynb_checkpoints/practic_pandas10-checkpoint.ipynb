```python
def who(x):
    if x < 18:
        return "Child"
    elif x > 18:
        return "Adult"
    elif x > 65:
        return "Old Ones"
```

```python
data["people"] = data["age"].apply(who)
```

```python
data
```

```python
len(data[data["people"] == "Child"])
```

```python
len(data[(data["survived"] == 1) & (data["people"] == "Child")])
```
