```python
df = sns.load_dataset("titanic")#from seborn import ddta set use function lode dataset
```

```python
df.head()
```

```python
df.iplot(x = "sex",y = "survived",kind = "bar")
```

```python
# check how many are survived
df.iplot(x = "sex",y = "survived",kind = "bar",xTitle = "sex",yTitle = "survived")
```

```python
# check relation between age and fare
df[["age","fare"]].iplot()
```

```python
df[["age","fare"]].iplot(x = "age",y = "fare",mode="markers",size=5)
```

```python
tip = sns.load_dataset("tips")
```

```python
tip
```

```python
# relation between tip and bill
tip[["total_bill","tip"]].iplot(x = "total_bill",y = "tip",mode="markers",size=5,xTitle="total_bill",yTitle="tip")
```

```python
# plot 3D graph  relation between tip and bill and size
tip.iplot(kind = "scatter3d",x = "total_bill",y = "tip",z = "size" )
```

```python
tip.iplot(kind = "bubble3d",x = "total_bill",y = "tip",z = "size",size = "size")
```

```python
tip
```

```python
# using seaborn relation in total bill and tips
sns.relplot(x = "total_bill",y = "tip",data = tip)#we can use relationplot function to see difference
```

```python
# saprating data based on smoker
sns.relplot(x = "total_bill",y = "tip",data = tip,hue = "smoker")#hue give the cetagory 
```

```python
# saprating data based on size
sns.relplot(x = "total_bill",y = "tip",data = tip,hue = "size")
```

```python
sns.relplot(x = "total_bill",y = "tip",data = tip,style = "size")#style parameter give style
```

```python
sns.relplot(x = "total_bill",y = "tip",data = tip,hue = "smoker",style = "time")
```
