```python
def exp(x):
    if x == "EN":
        return "Entry Level"
    elif x == "SE":
        return "Senior Level"
    elif x == "EX":
        return "Leader"
    else:
        return "Masters"
```

```python
df["experience"] = df["experience_level"].apply(exp)
```
